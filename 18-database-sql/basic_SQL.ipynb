```python
import pandas as pd
import sqlalchemy as sa
```

```python
pd.set_option('display.max_columns', 99)
```

```python
db_uri = "sqlite:///data/superstore.sqlite"
#or "postgresql://username@host/dbname" Also password 

engine = sa.create_engine(db_uri)
```

```python
# select random 5 rows
pd.read_sql('''
select * from superstore limit 5
''', engine)
```

```python
# select unique ship_mode
pd.read_sql('''
select distinct ship_mode from superstore
''', engine)
#equivalent of df.ship_mode.unique()
```

```python
# select unique country
pd.read_sql('''
select distinct country from superstore
''', engine)
```

```python
# number of rows by market
pd.read_sql('''
select market, count(*) num_rows from superstore
group by 1
''', engine)

#df.market.value_counts()
#df.groupby("market").size
```

```python
# number of rows by market, in descending order
pd.read_sql('''
select market, count(*) num_rows from superstore
group by 1
order by 2 desc
''', engine)

#group by 1: The first column 
#order by the count in descending order 
```

```python
# select market with number of rows > 6000
pd.read_sql('''
select market, count(*) num_rows from superstore
group by market
having num_rows > 6000
''', engine)
```

```python
# filter to market = Canada
pd.read_sql('''
select * from superstore
where market = 'Canada'
''', engine)
```

```python
# total profit by each category
pd.read_sql('''
select category, sum(profit) total_profit from superstore
group by 1
''', engine)
```

```python
# basic arithmetic in sql
pd.read_sql('''
select sales, profit, shipping_cost, (sales - profit) cost from superstore
limit 10
''', engine)


```

```python
# subquery (nested query) to get the country with maximum profit for each market
pd.read_sql('''
select market, country, profit from superstore
where (market, profit) in (
    select market, max(profit) max_profit from superstore
    group by 1
)
''', engine)
```

```python
# alternatively, do an inner join
pd.read_sql('''
select superstore.market, country, profit from superstore
inner join (
    select market, max(profit) max_profit from superstore
    group by 1
) market_profit on 
superstore.market = market_profit.market and superstore.profit = market_profit.max_profit
''', engine)
```

```python
# alternatively use a window / analytic function
pd.read_sql('''
with market_profit as (select market, country, profit,
rank() over (partition by market order by profit desc) as rank_profit
from superstore)
select market, country, profit from market_profit where rank_profit = 1
''', engine)
```

```python
# calculate the cost (sales - profit) per quantity for each transaction

#actually you just need select (sales-profit)/quantity, but you do "select sales, profit, quantity" to check if answers are correct. 

pd.read_sql('''
select sales, profit, quantity, ((sales - profit) / quantity) cost_per_quantity 
from superstore
''', engine)
```

```python
# average shipping cost per quantity by ship mode
pd.read_sql('''
select ship_mode, avg(shipping_cost / quantity) avg_shipping_cost_per_quantity 
from superstore
group by ship_mode
''', engine)




```

```python
# variance of sales by each subcategory

pd.read_sql('''
with summary as (
    select superstore.sub_category, (sales - avg_sales) * ?(sales - avg_sales) squared deviation, 
    count(*) as num_rows
    from superstore 
    inner join(
        select sub_category, avg(sales) avg_sales
        from superstore
        group by 1
    ) avg_sales_by_sub_category 
    on superstore.sub_category = avg_sales_by_sub_category.sub_category 
    group by 1 
)
select sub_category, (squared_deviation)/(num_rows - 1) as standard deviation
from summary
group by 1
''', engine)





```

```python
# city with highest total quantity by each region

pd.read_sql('''

with region_city_quantity as (
    select region, city, sum(quantity) as total_quantity from superstore)
    group by 1, 2), 
region_quantity as(
    select *, rank() over (partition by region order by total_quantity- desc) as rank_quantity
    from region_city_quantity)

select region, city, total_quantity from region_quantity where rank_quantity = 1

''', engine)
```

```python
# top 3 subcategories (by sales) by each category

pd.read_sql('''

with category_sub_sales as (
    select category, sub_category, sum(sales) as total_sales from superstore 
    group by 1, 2), 
category_sales as (
    select *, rank() over (partition by category order byt otal_sales desc) as rank_sales 
    from category_sub_sales)

select category, sub_category, rank_sales, total_sales from category_sales where rank_sales < 4
order by 1, 3 
''', engine)
```
